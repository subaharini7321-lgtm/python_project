```python
import numpy as np

a = np.array([88, 64, 72])
b = np.array([56, 69, 95])

print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
```

```python
#find maximum and minimum
import numpy as np
a=np.array([65,97,52])
b=np.array([78,56,56])
max=np.maximum(a,b)
print("maximum value:",max)
min=np.minimum(a,b)
print("minimum value:",min)
```
